```python
import math
import matplotlib.pyplot as plt
```

```python
for num in range(2, 12, 2):
    print(math.sin(1 + 3 ** num))
```

```python
sum = 0

for num in range(1, 13, 1):
    sum += 1 / (num ** 3)

print(sum)
```

```python
fig = plt.figure()
sp = fig.add_subplot(111)

for N in range(4, 13, 1):
    x = []
    y = []
    for n in range(N + 1):
        x.append(math.cos(2 * math.pi * n / N))
        y.append(math.sin(2 * math.pi * n / N))
    sp.plot(x, y)
    
sp.set_aspect('equal', adjustable='box')
plt.show()
```
