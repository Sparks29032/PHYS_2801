```python
#PART A)
#The small angle approximation for period of a pendulum is T = 2*pi*sqrt(l/g)
#In our case, l = 1 m and g = 9.81 m/s^2
#Plugging this in gives T = 2.006 s
```

```python
#PART B)

#define constants M = 1 kg, g = 9.81 m/s^2, l = 1 m
M = 1
g = 9.81
l = 1

#we know acceleration is -Mgsin(theta)
#So angular acceleration alpha = acceleration / l
def alpha (theta):
    acceleration = - M * g * math.sin(theta)
    return acceleration / l

#define initial conditions: initial angular velocity is (3 m/s) / l = 3 rad/s
#initial angle = 0 radians
omega_0 = 3 / l
omega = omega_0
theta_0 = 0
theta = theta_0

#we want time to increment from 0 to 10 with step size 0.001
t = 0
t_max = 10
t_inc = 0.001

#arrays where we'll store our times and thetas
times = []
positions = []

#stores the period
T = 0
flag = False

#find new angular velocity and angular position at each timestep
while t <= t_max:
    #add the times and positions to the arrays
    times.append(t)
    positions.append(theta)
    
    #increment angular position by angular velocity times the time increment
    theta += omega * t_inc
    
    #check if it's reached half a cycle
    if theta < 0 and flag == False:
        flag = True
        T = t * 2
    
    #increment angular velocity by angular acceleration times the time increment
    omega += alpha(theta) * t_inc
    
    #increment time
    t += t_inc
    
#plot it
import matplotlib.pyplot as plt
fig = plt.figure()
sp = fig.add_subplot(1, 1, 1)
sp.plot(times, positions, 'g-')
plt.title("Angular displacement over time from 0 <= t <= 10 seconds")
plt.xlabel("Time in seconds")
plt.ylabel("Angular displacement in radians")
plt.show(sp)

#the period of our non-approximated curve seems similar to the approximated period calculated above in part A)
print("The period is " + str(T) + " seconds which is close to the 2.006 seconds calculated in part A.")
```

```python
#PART C)
#Let us use X to represent theta and w represent omega = sqrt(g/l)
#The equation of motion is x(t) = Asin(wt) + Bcos(wt) for some A, B
#As we know X(0) = 0, x(0) = Asin(0) + Bcos(0) = B, so B = 0
#We also know X'(t) = Awcos(wt) = 3 when t = 0
#Thus, Aw = 3, so A = 3/w = 3sqrt(l/g)
#Plugging in g = 9.81 and l = 1 gives A = 0.958
#Thus, we should have X(t) = 0.958
#However, we can get a better amplitude since we know kinetic energy is maximized at theta = 0
#Thus KE = 0.5 Mv^2 = 0.5 * 1 kg * (3 m/s)^2 = 4.5 J
#At the top, we have Mg(l-lcos(X_max)) PE and no KE
#Thus, 1 kg * 9.81 m/s^2 (1 m - (1 m) * cos(X_max)) = KE = 4.5 J
#That solves to get 1 - cos(X_max) = 4.5 / 9.81, so cos(X_max) = 1 - 4.5 / 9.81
#Thus, X_max = arccos(1 - 4.5 / 9.81) = 0.999 radians
#So, let's write X(t) = 0.999 * sin(wt) where w = sqrt(g/l) for g = 9.81 m/s^2, l = 1 m
#Let's plot that on the graph

#first we should write the function X(t)
def X(t):
    return 0.999 * math.sin(math.sqrt(9.81 / 1) * t)

#now we find X for 0 <= t <= 10 and add them to a list
t = 0
times_2 = []
positions_2 = []
while t <= t_max:
    times_2.append(t)
    positions_2.append(X(t))
    t += t_inc

#now we plot it
fig = plt.figure()
sp = fig.add_subplot(1, 1, 1)
sp.plot(times, positions, 'g-')
sp.plot(times_2, positions_2, 'r-')
plt.title("Green: non-approximated, Red: small angle approximation")
plt.xlabel("Time in seconds")
plt.ylabel("Angular displacement in radians")
plt.show(sp)

#we see that the frequency of the small angle curve is slightly higher than that of the non-approximated curve
#this is because T_approx = 2.006 seconds while T_nonapprox = 2.120
#since frequency is inverse to period, as the non-approximated curve has a larger period, it has a smaller frequency
```
